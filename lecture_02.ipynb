```python
import sys
import matplotlib
matplotlib.use('Agg')

# import pandas as pd
import matplotlib.pyplot as plt

x =[1,2,3,4,6,7]
y =[2,3,4,5,6,7]

plt.bar(x,y)
plt.show()

plt.savefig("ouput.png")
sys.stdout.flush()
```
